```python
from bs4 import BeautifulSoup as bs
import requests
from splinter import Browser
import time

from zipfile import ZipFile

```

```python
executable_path = {'executable_path': 'C:/Users/jamie/Documents/chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://s3.amazonaws.com/tripdata/index.html'
```

```python
browser.visit(url)
```

```python
zip_url = []
zip_name = []
```

```python
html = browser.html
# Parse HTML with Beautiful Soup
soup = bs(html, 'html.parser')
# print(soup)
# Retrieve all elements that contain book information
links = soup.find_all('a')
print(links)
time.sleep(5)
```

```python

try:
    for link in links:
          
        
        a = link.get('href')
        b = a.rsplit('tripdata/', 1)[1]
        zip_url.append(a)    
        zip_name.append(b)
        


        
except:
    print("\nScraping Complete")
```

```python
zip_name
```

```python
del zip_url[-1]
del zip_name[-1]
```

```python
zip_url
```

```python
for x in range(len(zip_url)):
    headers = headers={'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Cafari/537.36'}
    
    url = zip_url[x]
    r = requests.get(url, allow_redirects=True, headers=headers)
    open(zip_name[x], 'wb').write(r.content)
    

    zf = ZipFile(zip_name[x])
    zf.extractall(path = '/data/')
    # close the ZipFile instance
    zf.close()
    
```

```python
for x in range(len(zip_url)):
    with ZipFile(zip_name[x], 'r') as zipObj:
   # Extract all the contents of zip file in current directory
       zipObj.extractall('data')
```

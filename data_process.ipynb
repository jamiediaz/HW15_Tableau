```python
import csv
import numpy as np
import pandas as pd
import requests
from os import listdir
from os.path import isfile, join
```

```python
pd.set_option('display.max_columns', None)
```

```python
file_names = [f for f in listdir('data') if isfile(join('data', f))]
```

```python
file_names
```

```python

file_names = [ x for x in file_names if "JC-" not in x ]
bike_dict = {}
bike_dict['2013'] = [ x for x in file_names if "2013" in x ]
bike_dict['2014'] = [ x for x in file_names if "2014" in x ]
bike_dict['2015'] = [ x for x in file_names if "2015" in x ]
bike_dict['2016'] = [ x for x in file_names if "2016" in x ]
bike_dict['2017'] = [ x for x in file_names if "2017" in x ]
bike_dict['2018'] = [ x for x in file_names if "2018" in x ]
bike_dict['2019'] = [ x for x in file_names if "2019" in x ]
```

```python
yr_csv = f"data/{bike_dict['2013'][0]}"
bike_df = pd.read_csv(yr_csv, encoding = "ISO-8859-1")
bike_df.columns = map(str.lower, bike_df.columns)
bike_df.columns = bike_df.columns.str.replace(' ', '')

for x in bike_dict['2013'][1:]: 
    incoming_csv = f"data/{x}"
    print (incoming_csv)
    incoming_df = pd.read_csv(incoming_csv, encoding = "ISO-8859-1")
    incoming_df.columns = map(str.lower, incoming_df.columns)
    incoming_df.columns = incoming_df.columns.str.replace(' ', '')
    bike_df = bike_df.append(incoming_df, ignore_index=True, sort=False)

print(bike_df.count())
print(bike_df.shape)
bike_df.to_csv("tab_data/bike_2013.csv", index=False, header=True)
bike_df.head(1)
```
